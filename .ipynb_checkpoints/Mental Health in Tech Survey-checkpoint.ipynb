```python
### SO what are the companyes how provide more care programes for M ill people!



### DS_ILL ####

care = ds_ill.groupby(['no_employees']).agg({'benefits': 'value_counts','seek_help': 'value_counts', 'anonymity':'value_counts' })

care.index.set_names(['employees_numb', 'answer'], inplace=True)

## Total diferent  answers by company size ##

care_total = pd.DataFrame(care.sum(axis = 1), columns = ['total'])

care_total.reset_index(level = 'answer', inplace=True)

care_total.reset_index(level = 'employees_numb', inplace=True)

##

## Total of answers by company size ##

sum_cond_by_size = care_total.groupby(['employees_numb']).agg({'total':'sum'})

sum_cond_by_size.reset_index(level = 'employees_numb', inplace=True)

##


##  Relative Frquency of total don't know answers by company size  ##

unkonwn_by_size = care_total[care_total['answer'] == "Don't know"]

unkonwn_by_size = unkonwn_by_size.reset_index().drop(columns = ['index', 'answer' ])

unkonwn_by_size['total'] = unkonwn_by_size['total']/sum_cond_by_size['total']

## 

care
```

```python
##  Relative Frquency of total 'No' answers by company size  ##

no_by_size = care_total[care_total['answer'] == "No"]

no_by_size = no_by_size.reset_index().drop(columns = ['index', 'answer' ])

no_by_size['total'] = no_by_size['total']/sum_cond_by_size['total']
no_by_size
```

```python
##  Relative Frquency of total 'Yes' answers by company size  ##



yes_by_size = care_total[care_total['answer'] == "Yes"]

yes_by_size = yes_by_size.reset_index().drop(columns = ['index', 'answer'])

yes_by_size['total'] = yes_by_size['total']/sum_cond_by_size['total']
yes_by_size
```

```python
##Correlation##

print(no_by_size.corr()) ## high correlation

print(yes_by_size.corr()) ## high inverse correlation

print(unkonwn_by_size.corr()) ##no correlation 
```
